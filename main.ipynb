```python
#IMPORT DATASET USING SpaceX API

import requests
import pandas as pd
import matplotlib.pyplot as plt

url = "https://api.spacexdata.com/v4/launches"

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    df = pd.json_normalize(data)
    print('Data Extraction Successfull')
else:
    print('Unable to Fetch Data, Try Again Later')
```

```python
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 50)
df
```

```python
df2 = df[['rocket', 'success', 'details', 'payloads',	'launchpad', 'flight_number', 'name', 'date_utc', 'id', 'fairings.reused', 'fairings.recovery_attempt',	'fairings.recovered']]
```

```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)
df2
```

```python

df2[['name', 'success', 'date_utc']]
```

```python
df2['date_utc'] = pd.to_datetime(df2['date_utc'])
df2['date_utc'] = df2['date_utc'].dt.year
df2
```

```python
#ANALYZING SUCCESSFUL LAUNCHES OVER YEARS
```

```python
df_succ = df2[df2['success'] == True]
df_succ
```

```python
df_yc = df_succ['date_utc'].value_counts().sort_index(ascending=True)
df_yc
```

```python
df_yc.plot(kind='line', title='Successful Launch Trends', figsize=(16,9), xlabel='Year', ylabel='Launches')
plt.savefig('plots/successful_launch_trends.png', dpi=300, bbox_inches='tight')
```

```python
df_recatt = df2[df2['fairings.recovery_attempt'] == True]

#THIS IS ANOTHER METHOD ONE CAN USE TO FILTER DATA
#df2[~df2['fairings.recovered'].isin([True, False])][['fairings.recovered']]
```

```python
df_recatt[['date_utc', 'fairings.recovery_attempt', 'fairings.recovered']]
df_recatt[df_recatt['fairings.recovered'] == True]
```

```python
df_recatt_sucess = df_recatt[df_recatt['fairings.recovered'] == True].groupby('date_utc').size()
df_recatt_sucess
```

```python
df_rr = (df_recatt_sucess/df_recatt_attempts * 100).fillna(0)
df_rr
```

```python
df_final = pd.DataFrame({
    'Attempts': df_recatt_attempts,
    'Successes': df_recatt_sucess
}).fillna(0).astype(int)

df_final
```

```python
df_final.plot(kind = 'bar', figsize= (16,9), title= 'Bar Chart of Attempts Vs Successful', xlim=(2017, 2021))
plt.savefig('plots/fairings_recovery_attemps_successes', dpi = 300, bbox_inches = 'tight')
```

```python
df_rr.plot(kind='line', title='Fairing Recovery Success Rate (%) Over Time', ylim=(0, 100), xlim=(2017, 2021), figsize=(16,9), xlabel='Years', ylabel='Success Rate')
plt.savefig('plots/fairings_recover_rate_over_time', dpi = 300, bbox_inches = 'tight')
```

```python
#TASK 2 - IMPORTING LAUNCHPADS DATA FROM API AND DOING EXPLANATORY DATA ANALYSIS WITH THAT.
```

```python
import requests
import pandas as pd

url = "https://api.spacexdata.com/v4/launchpads"

response  = requests.get(url)

if response.status_code == 200:
    launchpads = response.json()
    df_lp = pd.json_normalize(launchpads)
    print('Launchpad Data Extracted Sucessfully.', 'Status Code:', response.status_code)
else:
    print("Couldn't Get Data")
```

```python
pd.set_option('display.max_rows', 5)
pd.set_option('display.max_columns', 500)

#USED BELOW FILTER TO ONLY SHOW USEFUL ROWS BUT REALISED LATER THAT I CAN DO THAT AFTER JOINING THE TWO DATASETS
#df_lp2 = df_lp[['name', 'locality', 'region', 'launch_attempts', 'launch_successes', 'rockets', 'status', 'id']]
df_lp
```

```python
df2
```

```python
df_lpjoin = df_lp.merge(df2, how='inner', left_on = 'id', right_on='launchpad')
df_lpjoin
```

```python
# EDA ABOUT LAUNCHPAD SUCCESS RATES
```

```python
df_lpjoin_lpscr = df_lpjoin.dropna(subset=('success'))
df_lpjoin_lpscr['success'] = df_lpjoin_lpscr['success'].astype(int)
df_lpjoin_lpscr
```

```python
df_lpjoin_lpscr_final  = df_lpjoin_lpscr.groupby('name_x')['success'].mean() * 100
df_lpjoin_lpscr_final
```

```python
#EDA ABOUT TOTAL LAUNCHES FROM EACH LAUNCHPADS
```

```python
df_lpla = df_lp[['name','launch_attempts','launch_successes']]
df_lpla.set_index('name', inplace=True)
df_lpla
```

```python
# PLOTTING THE GRAPHS
```

```python

import seaborn as sns

# 1. Map the First Plot
sns.set_theme(style='whitegrid')
plt.figure(figsize=(16, 9))

ax = sns.barplot(x=df_lpjoin_lpscr_final.index, y=df_lpjoin_lpscr_final.values)
ax.set_title('Calculated Success Rate by Site (Verified via launch Logs)', fontsize=15)
ax.set_xlabel('Launchpad Code Names')
ax.set_ylabel('Success Rate (%)')
plt.savefig('plots/calculated_success_rate', dpi = 300, bbox_inches = 'tight')
plt.show()
# YOU NEED TO CALL plt.savefig BEFORE plt.show CUS plt.show CLAERS THE FIG FROM MEMORY AND YOU'LL GET A BLANK IMAGE
```
